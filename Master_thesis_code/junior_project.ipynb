```python
from glob import glob
from tqdm import tqdm
import os
from path import Path
import pandas as pd
import numpy as np
from keras.utils import to_categorical
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
obj1_files = glob('Desktop\Files for task 1 and task 6\Object 1\*')

df = pd.DataFrame()
for i in tqdm(obj1_files):
    df = pd.concat([df, pd.read_excel(i, header=None).iloc[:,6:]])
```

```python
df['label'] = 1
df.shape
df.head()
```

```python
obj2_files = glob('Desktop\Files for task 1 and task 6\Object 2\*')

df_2 = pd.DataFrame()
for i in tqdm(obj2_files):
    df_2 = pd.concat([df_2, pd.read_excel(i, header=None).iloc[:,6:]])
```

```python
df_2['label'] = 0
df_2.shape
df_2.head()
```

```python
## Merge two data sets

merge_df = pd.concat([df, df_2])
merge_df
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(merge_df.iloc[:,:3400], merge_df['label'], test_size=0.2)
```

```python
# number of classes in binary form  

# Find the unique numbers from the train labels
classes = np.unique(y_train)
num_classes = len(classes)
print('Total number of outputs : ', num_classes)
print('Output classes : ', classes)
```

# CNN Model

```python
from sklearn.model_selection import train_test_split

from sklearn.preprocessing import LabelEncoder
import numpy as np
import keras
from keras.models import Sequential
from keras.layers import Dense, Flatten, Conv1D, MaxPool1D, Dropout
 
import matplotlib.pyplot as plt
```

```python
# one hot encoding
y_train_new = keras.utils.to_categorical(y_train, 2)
y_test_new = keras.utils.to_categorical(y_test, 2)
#print(y_train_new)


# reshape 2D to 3D
X_train_new = X_train.values.reshape(412, 3400, 1)
X_test_new = X_test.values.reshape(103, 3400, 1)


```

```python
# build CNN model
model = Sequential()
model.add(Conv1D(64,2, input_shape=(3400, 1), activation='relu'))  # convolation
model.add(MaxPool1D(pool_size=2))  # pooling
model.add(Flatten())  # flatten
model.add(Dense(128, activation='relu'))  # fc
model.add(Dropout(0.3))  # dropout
model.add(Dense(2, activation='softmax'))
 
# model compile
model.compile(loss=keras.losses.categorical_crossentropy,
              optimizer=keras.optimizers.Adadelta(),
              metrics=['accuracy'])
model.summary()

```

```python
batch_size = 128
epochs = 70
model2 = model.fit(X_train_new, y_train_new,
                  batch_size=batch_size,
                  epochs=epochs,
                  verbose=2,
                  validation_data=(X_test_new, y_test_new))
 
# draw loss
plt.plot(model2.history['loss'])
plt.plot(model2.history['val_loss'])
plt.title('model train vs validation loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'validation'], loc='upper right')
plt.show()
```

```python
predicted_classes = model.predict(X_test_new)
predicted_classes1 = predicted_classes.argmax(axis=1)
predicted_classes1
```

```python

from sklearn.metrics import classification_report

print(classification_report(y_test, predicted_classes1))
```

```python
# find the TP, TN, FP, FN,FDR, NPV, TPR,TNR


def print_confusion_matrix(y_true, y_pred):
    cm = confusion_matrix(y_true, y_pred)
    TP = cm[0][0]
    FP = cm[0][1]
    FN = cm[1][0]
    TN = cm[1][1]
    print('True positive = ', cm[0][0])
    print('False positive = ', cm[0][1])
    print('False negative = ', cm[1][0])
    print('True negative = ', cm[1][1])
    
    # Sensitivity, hit rate, recall, or true positive rate
    TPR = TP/(TP+FN)
    # Specificity or true negative rate
    TNR = TN/(TN+FP) 
    # Precision or positive predictive value
    PPV = TP/(TP+FP)
    # Negative predictive value
    NPV = TN/(TN+FN)
    # Fall out or false positive rate
    FPR = FP/(FP+TN)
    # False negative rate
    FNR = FN/(TP+FN)
    # False discovery rate
    FDR = FP/(TP+FP)
    # Overall accuracy
    ACC = (TP+TN)/(TP+FP+FN+TN)
    
    print('TPR:', TPR )
    print('TNR:', TNR )
    print('PPV:', PPV )
    print('NPV:', NPV )
    print('FPR:', FPR )
    print('FNR:', FNR )
    print('FDR:', FDR )
    print('ACC:', ACC )
    
    return TPR, FPR
    
TPR, FPR = print_confusion_matrix(y_test, predicted_classes1)
```
