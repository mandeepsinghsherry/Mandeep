```python
# Pythono3 code to rename multiple 
# files in a directory or folder 

# importing os module 
import os 
import re

# Function to rename multiple files 
def main(): 
    path = "c://Users/MANDEEP SINGH SHERRY/Desktop/fun/"
    x= os.listdir(path)
    x.sort(key=lambda f: int(re.sub('\D', '', f)))
    
    for count, filename in enumerate(x): 
        print(filename)
        dst ="chair_bottle" + str(count) + ".jpg"
        src = path+ filename 
        dst = path+ dst 

        # rename() function will 
        # rename all the files 
        os.rename(src, dst) 

# Driver Code 
if __name__ == '__main__': 

# Calling main() function 
 main() 
```
