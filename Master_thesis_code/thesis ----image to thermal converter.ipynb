```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
import glob
import cv2
import timeit
from PIL import Image
import re



numbers = re.compile(r'(\d+)')
def numericalSort(value):
    parts = numbers.split(value)
    parts[1::2] = map(int, parts[1::2])
    return parts

# img_src = [cv2.imread(file) for file in glob.glob("C:/Users/MANDEEP SINGH SHERRY/Desktop/sensor fusion/bird test/*.jpg")]
# img_src

#img_src = Image.open('C:/Users/MANDEEP SINGH SHERRY/Desktop/sensor fusion/bird test/bt1.jpg').convert('L')
image_list = []
def my_load_Train_data_2() :
    i = 0

    for filename in sorted(glob.glob('c://Users/MANDEEP SINGH SHERRY/Desktop/Thesis Research/Two models/Camera Images/testing/class 1/*.jpg'),key=numericalSort): #assuming gif
        img_src=Image.open(filename).convert('L')
        cm_hot = mpl.cm.get_cmap('jet')
        img_src.thumbnail((512,512))
        im = np.array(img_src)
        im = cm_hot(im)
        im = np.uint8(im * 255)
        im = Image.fromarray(im).convert('RGB')
        image_list.append(im)
        im.save('C:/Users/MANDEEP SINGH SHERRY/Desktop/Thesis Research/Two models/Thermal Images/testing/class 1/thermal_bed'+str(i)+'.jpg')
        i=i+1

    return image_list


my_load_Train_data_2()

```
