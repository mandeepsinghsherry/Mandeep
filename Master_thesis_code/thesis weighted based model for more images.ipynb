```python
from PIL import Image
import glob
import tensorflow as tf
from keras.preprocessing.image import load_img
from keras.preprocessing.image import img_to_array
import cv2 as cv
import re
import os
```

```python
numbers = re.compile(r'(\d+)')
def numericalSort(value):
    parts = numbers.split(value)
    parts[1::2] = map(int, parts[1::2])
    return parts

def loadWeightedAverageCameraImages() :
    image_list1 = []
    for filename in sorted(glob.glob('C:/Users/MANDEEP SINGH SHERRY/Desktop/Thesis Research/Two models/brovey_output/training/table_specs/*.jpg'),key=numericalSort): #assuming gif
         image_list1.append(filename)
    return image_list1


def loadWeightedAverageThermalImages() :
    image_list2 = []
    for filename in sorted(glob.glob('C:/Users/MANDEEP SINGH SHERRY/Desktop/Thesis Research/Two models/weighted/training/table_specs/*.jpg'),key=numericalSort): #assuming gif
        image_list2.append(filename)
    return image_list2
    


WI1= loadWeightedAverageCameraImages()
WI2= loadWeightedAverageThermalImages()

for j in range(len(WI1)):
    img1 = cv.imread(WI1[j],0)
    img2 = cv.imread(WI2[j],0)
    dst1=cv.addWeighted(img1,0.50,img2,0.50,0) #Weighted fusion processing means that picture 1 accounts for 25% and picture 2 accounts for 75%
    cv.imwrite("C:/Users/MANDEEP SINGH SHERRY/Desktop/Thesis Research/Two models/combined/training/table_specs/table_specs_WGH_BRV"+str(j)+".jpg",dst1)
    print('fusion done')


f = r'c://Users/MANDEEP SINGH SHERRY//Desktop/Thesis Research/Two models/combined/training/table_specs'

for file in os.listdir(f):
    f_img = f+"/"+file
    img = Image.open(f_img).convert('RGB')
    img = img.resize((225,225))
    img.save(f_img)
```
